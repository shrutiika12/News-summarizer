```python
from gnews import GNews

def fetch_news(topic, num_articles=5):
    """
    Fetches news articles on a given topic.
    Uses GNews description directly — no scraping needed.
    """
    google_news = GNews(language='en', country='IN', max_results=num_articles)
    raw_articles = google_news.get_news(topic)
    
    articles = []
    
    for item in raw_articles:
        try:
            title = item.get('title', '')
            description = item.get('description', '')
            
            # combine title + description for more text
            full_text = title + ". " + description
            
            if len(full_text) > 30:
                articles.append({
                    'title': title,
                    'text': full_text,
                    'url': item.get('url', ''),
                    'source': item['publisher']['title']
                })
                print(f"✓ Fetched: {title[:60]}...")
                
        except Exception as e:
            print(f"✗ Skipped: {e}")
    
    print(f"\nSuccessfully fetched {len(articles)} articles!")
    return articles


# Test
topic = input("Enter a topic to search news for: ")
articles = fetch_news(topic, num_articles=5)
```

```python
# Summarization
from transformers import BartForConditionalGeneration, BartTokenizer

print("Loading summarization model...")

model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

print("✓ Model loaded successfully!")
```

```python
def chunk_text(text, max_words=700):
    words = text.split()
    chunks = []
    for i in range(0, len(words), max_words):
        chunk = ' '.join(words[i:i + max_words])
        chunks.append(chunk)
    return chunks
```

```python
def summarize_article(text, max_length=130, min_length=30):
    # chunk if too long
    words = text.split()
    if len(words) > 700:
        text = ' '.join(words[:700])
    
    inputs = tokenizer(
        text,
        return_tensors="pt",
        max_length=1024,
        truncation=True
    )
    
    summary_ids = model.generate(
        inputs["input_ids"],
        max_length=max_length,
        min_length=min_length,
        length_penalty=2.0,
        num_beams=4,
        early_stopping=True
    )
    
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary


# Test it!
summary = summarize_article(articles[0]['text'])
print("TITLE:", articles[0]['title'])
print("\nSUMMARY:", summary)
```

```python
def summarize_all(articles):
    results = []
    
    for i, article in enumerate(articles):
        print(f"Summarizing article {i+1}/{len(articles)}: {article['title'][:50]}...")
        
        summary = summarize_article(article['text'])
        
        results.append({
            'title': article['title'],
            'source': article['source'],
            'url': article['url'],
            'original_text': article['text'],
            'summary': summary
        })
        
        print(f"✓ Done!\n")
    
    return results

topic = input("Enter a topic to search news for: ")
articles = fetch_news(topic, num_articles=5)
summarized_articles = summarize_all(articles)


for item in summarized_articles:
    print("=" * 60)
    print("TITLE:", item['title'])
    print("SUMMARY:", item['summary'])
    print()
```

```python
import random
import re
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.tag import pos_tag

def generate_mcqs(summary, num_questions=3):
    """
    Generates MCQ questions from a summary.
    """
    stop_words = set(stopwords.words('english'))
    sentences = sent_tokenize(summary)
    
    if len(sentences) == 0:
        return []
    
    mcqs = []
    used_answers = set()
    
    for sentence in sentences[:num_questions]:
        # tag words with their part of speech
        words = word_tokenize(sentence)
        tagged = pos_tag(words)
        
        # pick nouns or numbers as the answer (most meaningful words)
        candidates = [
            word for word, tag in tagged
            if tag in ('NN', 'NNP', 'NNS', 'NNPS', 'CD')
            and word.lower() not in stop_words
            and len(word) > 2
            and word not in used_answers
        ]
        
        if not candidates:
            continue
        
        # pick the answer word
        answer = random.choice(candidates)
        used_answers.add(answer)
        
        # create the question by blanking the answer
        question = sentence.replace(answer, "______")
        
        # generate 3 wrong options from other sentences
        all_words = word_tokenize(summary)
        all_tagged = pos_tag(all_words)
        wrong_pool = [
            word for word, tag in all_tagged
            if tag in ('NN', 'NNP', 'NNS', 'NNPS', 'CD')
            and word.lower() not in stop_words
            and len(word) > 2
            and word != answer
        ]
        
        # make sure we have enough wrong options
        wrong_pool = list(set(wrong_pool))
        if len(wrong_pool) < 3:
            continue
            
        wrong_options = random.sample(wrong_pool, 3)
        
        # combine answer + wrong options and shuffle
        options = wrong_options + [answer]
        random.shuffle(options)
        
        mcqs.append({
            'question': question,
            'options': options,
            'answer': answer
        })
    
    return mcqs


# Test on first summarized article
mcqs = generate_mcqs(summarized_articles[0]['summary'], num_questions=3)

print("SUMMARY USED:")
print(summarized_articles[0]['summary'])
print("\n" + "="*60)
print("GENERATED MCQs:")
print("="*60)

for i, mcq in enumerate(mcqs):
    print(f"\nQ{i+1}: {mcq['question']}")
    for j, option in enumerate(mcq['options']):
        print(f"   {chr(65+j)}) {option}")
    print(f"   ✓ Answer: {mcq['answer']}")
```

```python
def add_mcqs_to_all(summarized_articles, num_questions=3):
    """
    Adds MCQs to every summarized article.
    """
    final_results = []
    
    for i, article in enumerate(summarized_articles):
        print(f"Generating MCQs for article {i+1}/{len(summarized_articles)}...")
        
        mcqs = generate_mcqs(article['summary'], num_questions=num_questions)
        
        final_results.append({
            'title': article['title'],
            'source': article['source'],
            'url': article['url'],
            'original_text': article['original_text'],
            'summary': article['summary'],
            'mcqs': mcqs
        })
        
        print(f"✓ {len(mcqs)} MCQs generated!\n")
    
    return final_results


# Run it!
final_results = add_mcqs_to_all(summarized_articles)
```

```python
from rouge_score import rouge_scorer

print("ROUGE scorer ready!")
```

```python
def calculate_rouge(generated_summary, reference_summary):
    """
    Calculates ROUGE-1, ROUGE-2 and ROUGE-L scores.
    ROUGE-1 = single word overlap
    ROUGE-2 = two word pair overlap  
    ROUGE-L = longest common sequence
    """
    scorer = rouge_scorer.RougeScorer(
        ['rouge1', 'rouge2', 'rougeL'],
        use_stemmer=True
    )
    
    scores = scorer.score(reference_summary, generated_summary)
    
    return {
        'rouge1': round(scores['rouge1'].fmeasure, 4),
        'rouge2': round(scores['rouge2'].fmeasure, 4),
        'rougeL': round(scores['rougeL'].fmeasure, 4)
    }


# Test on first article
# We use the original text as reference and AI summary as generated
reference = summarized_articles[0]['original_text']
generated = summarized_articles[0]['summary']

scores = calculate_rouge(generated, reference)

print("ROUGE Scores for Article 1:")
print(f"  ROUGE-1 (word overlap):      {scores['rouge1']}")
print(f"  ROUGE-2 (phrase overlap):    {scores['rouge2']}")
print(f"  ROUGE-L (sequence overlap):  {scores['rougeL']}")
```

```python
def evaluate_all(summarized_articles):
    """
    Calculates ROUGE scores for all articles
    and gives an overall average — this is your resume number!
    """
    all_scores = []
    
    for i, article in enumerate(summarized_articles):
        scores = calculate_rouge(
            article['summary'],
            article['original_text']
        )
        
        all_scores.append(scores)
        
        print(f"Article {i+1}: {article['title'][:45]}...")
        print(f"  ROUGE-1: {scores['rouge1']}  |  ROUGE-2: {scores['rouge2']}  |  ROUGE-L: {scores['rougeL']}")
        print()
    
    # calculate averages
    avg_rouge1 = round(sum(s['rouge1'] for s in all_scores) / len(all_scores), 4)
    avg_rouge2 = round(sum(s['rouge2'] for s in all_scores) / len(all_scores), 4)
    avg_rougeL = round(sum(s['rougeL'] for s in all_scores) / len(all_scores), 4)
    
    print("="*50)
    print("OVERALL AVERAGE SCORES:")
    print(f"  ROUGE-1: {avg_rouge1}")
    print(f"  ROUGE-2: {avg_rouge2}")
    print(f"  ROUGE-L: {avg_rougeL}")
    print("="*50)
    print(f"\nYour resume line:")
    print(f'  "Achieving summarization accuracy of ROUGE score {avg_rouge1} across articles processed"')
    
    return all_scores


# Run it!
all_scores = evaluate_all(summarized_articles)
```

```python
def run_full_pipeline(topic, num_articles=5):
    """
    Full pipeline:
    1. Fetch news
    2. Summarize
    3. Generate MCQs
    4. Calculate ROUGE
    """
    print(f"\n{'='*60}")
    print(f"Running full pipeline for topic: '{topic}'")
    print(f"{'='*60}\n")
    
    # Step 1 - fetch
    print("📰 Fetching news...")
    articles = fetch_news(topic, num_articles=num_articles)
    
    if not articles:
        print("No articles found. Try a different topic!")
        return
    
    # Step 2 - summarize
    print("\n🤖 Summarizing articles...")
    summarized = summarize_all(articles)
    
    # Step 3 - MCQs
    print("\n📝 Generating MCQs...")
    final = add_mcqs_to_all(summarized)
    
    # Step 4 - ROUGE
    print("\n📊 Calculating ROUGE scores...")
    scores = evaluate_all(summarized)
    
    # Final output
    print("\n\n" + "="*60)
    print("FINAL RESULTS")
    print("="*60)
    
    for i, article in enumerate(final):
        print(f"\n📌 ARTICLE {i+1}: {article['title'][:55]}...")
        print(f"   Source: {article['source']}")
        print(f"\n📄 SUMMARY:\n   {article['summary']}")
        print(f"\n❓ MCQs:")
        
        if not article['mcqs']:
            print("   No MCQs generated.")
        else:
            for j, mcq in enumerate(article['mcqs']):
                print(f"\n   Q{j+1}: {mcq['question']}")
                for k, option in enumerate(mcq['options']):
                    marker = "✓" if option == mcq['answer'] else " "
                    print(f"      {marker} {chr(65+k)}) {option}")
        
        print(f"\n   ROUGE-1: {scores[i]['rouge1']} | ROUGE-2: {scores[i]['rouge2']} | ROUGE-L: {scores[i]['rougeL']}")
        print("\n" + "-"*60)
    
    return final


# Run everything!
topic = input("Enter topic: ")
results = run_full_pipeline(topic)
```
